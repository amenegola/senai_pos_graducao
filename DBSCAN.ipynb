```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import pandas as pd
%matplotlib notebook
```

```python
iris = datasets.load_iris()
X = iris.data[:,[2,3]]
clustering = DBSCAN(eps=0.4, min_samples=10).fit(X)
X = pd.DataFrame(X)
Base = pd.DataFrame(clustering.labels_)
Base = pd.concat([Base,X], axis =1)
Base.columns = ['Label', 'X','Y']
plt.scatter(Base['X'],Base['Y'], c=Base['Label'])
```

```python
%matplotlib notebook

from mpl_toolkits.mplot3d import axes3d, Axes3D

fig = plt.figure()
ax = Axes3D(fig)

X = iris.data[:,[0,1,2]]
clustering = DBSCAN(eps=0.6, min_samples=20).fit(X)

ax.scatter(iris.data[:,1], iris.data[:,2], iris.data[:,3],c=clustering.labels_)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
